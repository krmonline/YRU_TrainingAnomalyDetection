```
import numpy as np
import h5py
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```
cpu = pd.read_csv("https://raw.githubusercontent.com/numenta/NAB/master/data/realAWSCloudwatch/ec2_cpu_utilization_53ea38.csv")
```

```
cpu2 = cpu.value.values.reshape(-1,288)
```

```
X = cpu2[4]
Y = cpu2[5]
clustering = DBSCAN(eps=0.05, min_samples=2).fit(np.asarray([X,Y]).T)
```

```
plt.figure(figsize=(20, 3))
plt.plot(X)
plt.plot(Y)
```

```
plt.figure(figsize=(10, 3))
plt.subplot(121).scatter(X,Y)
plt.subplot(122).scatter(X,Y,c=clustering.labels_)
```

```
plt.figure(figsize=(20, 3))
plt.plot(X)
plt.plot(Y)
for  i in range(len(clustering.labels_)):
  if clustering.labels_[i] == -1:
    plt.plot([i,i],[1.5,2.6],'r-')
```
